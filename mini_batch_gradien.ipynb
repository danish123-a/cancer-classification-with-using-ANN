```python
import numpy as np
```

```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
v=pd.read_csv("https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv")
```

```python
v
```

```python
v=v.drop(columns=["id","Unnamed: 32"])
```

```python
v
```

```python
x_train,x_test,y_train,y_test=train_test_split(v.iloc[:,1:],v.iloc[:,0],test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
std=StandardScaler()
```

```python
x_train,x_test=std.fit_transform(x_train),std.fit_transform(x_test)
```

```python
label=LabelEncoder()
```

```python
y_train,y_test=label.fit_transform(y_train),label.fit_transform(y_test)
```

```python
y_train
```

```python
import torch as t
```

```python
x_train_t,x_test_t,y_train_t,y_test_t=t.from_numpy(x_train),t.from_numpy(x_test),t.from_numpy(y_train),t.from_numpy(y_test)
```

```python
x_train_t
```

```python
import torch.nn as nn
```

```python
class Simple_nn(nn.Module):
  def __init__(self,features):
    super().__init__()
    self.container=nn.Sequential(nn.Linear(features,1),nn.Sigmoid())
  def forward(self,x):
    z=self.container(x)
    return z
```

```python
x_train_t,x_test_t,y_train_t,y_test_t=x_train_t.float(),x_test_t.float(),y_train_t.float(),y_test_t.float()
```

```python
model=Simple_nn(x_train_t.shape[1])
```

```python
model(x_train_t)
```

```python
loss1=nn.BCELoss()
```

```python
#ptimmizer
optimizer=t.optim.SGD(model.parameters(),lr=0.01)
```

```python
epochs=25
```

```python
batch_size=32
```

```python
for i in range(epochs):
  for start_index in range(0,len(x_train_t),batch_size):
    end_index=start_index+batch_size
    x_batch=x_train_t[start_index:end_index]
    y_batch=y_train_t[start_index:end_index]
    #forward_pass
    y_pred=model(x_train_t)
    #loss
    loss_=loss1(y_pred,y_train_t.reshape(-1,1))
    #clear gradient
    optimizer.zero_grad()
    #backward()
    loss_.backward()
    #update weights
    optimizer.step()
    print(loss_)
```

```python
model(x_train_t)
```

```python
#evaluation
with t.no_grad():
  y_pred=model(x_test_t)
  y_pred=(y_pred>0.5).int()
```

```python
y_pred
```

```python
#accuracy
accuracy_=((y_pred==y_test_t).float()).mean()
```
